```python
import pandas as pd
import networkx as nx
import numpy as np
G = nx.DiGraph()


G.add_node("Lung_cancer")
G.add_node("Smoking")
G.add_node("Yellow_Fingers")
G.add_node("Anxiety")
G.add_node("Peer_Pressure")
G.add_node("Genetics")
G.add_node("Attention_Disorder")
G.add_node("Born_an_Even_Day")
G.add_node("Car_Accident")
G.add_node("Fatigue")
G.add_node("Allergy")
G.add_node("Coughing")


G.add_edge("Coughing", "Fatigue")
G.add_edge("Allergy", "Coughing")
G.add_edge("Fatigue", "Car_Accident")
G.add_edge("Lung_cancer", "Coughing")
G.add_edge("Lung_cancer", "Fatigue")
G.add_edge("Attention_Disorder", "Car_Accident")
G.add_edge("Genetics", "Lung_cancer")
G.add_edge("Genetics", "Attention_Disorder")
G.add_edge("Smoking", "Lung_cancer")
G.add_edge("Smoking", "Yellow_Fingers")
G.add_edge("Peer_Pressure", "Smoking")
G.add_edge("Anxiety", "Smoking")

pos = {
    "Anxiety": (0, 0),
    "Peer_Pressure": (1, 0),
    "Born_an_Even_Day": (2, 0),
    "Yellow_Fingers": (0, 1),
    "Smoking": (1, 1),
    "Genetics": (2.3, 1),
    "Allergy": (-0.2, 2),
    "Lung_cancer": (1, 2),
    "Attention_Disorder": (2, 2),
    "Coughing": (0, 3),
    "Fatigue": (1, 3),
    "Car_Accident": (1.5, 4),
}
colors = ["cyan"] * 12
colors[0] = "r"
colors[8] = "r"
colors[1] = "g"

edge_colors = ["black"] * 12
# edge_colors[0] = "r"
# edge_colors[1] = "r"
# edge_colors[9] = "r"
# edge_colors[11] = "r"



nx.draw(G, pos= pos, with_labels=True, node_size=500, edgecolors = colors, node_color=colors, font_size=14,font_weight=700, edge_color=edge_colors)


nx.write_gml(G, "lucas_graph.gml")
```

```python
G.add_node("Driving_one_handed")

G.add_edge("Smoking", "Driving_one_handed")
G.add_edge("Driving_one_handed", "Car_Accident")

pos["Driving_one_handed"] = (1.6, 2.5)

nx.draw(G, pos= pos, with_labels=True, node_size=500, edgecolors = colors, node_color=colors + ["cyan"], font_size=14,font_weight=700, edge_color=edge_colors + ["black"])
```
